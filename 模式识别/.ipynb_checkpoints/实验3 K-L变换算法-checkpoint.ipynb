# 实验3 K-L变换算法

```python
import numpy as np
```

实验样本

```python
x = [-1, -5/-4, -4/-5, -5/-6, -6/-5, 1, 5/6, 6/5, 5/4, 4/5]
n = 10
```

## （1）计算样本向量的均值

```python
def get_u(x, n):
    """
    计算样本向量的均值
    x 样本向量/数组
    n 样本个数/阶数
    return 均值
    """
    SUM = 0;
    for i in range(n):
        SUM += x[i]
    return SUM/n
```

```python
u = get_u(x, n)
print("样本均值为：", u)
```

##       计算协方差矩阵

```python
def get_C(x, n, u):
    """
    计算协方差矩阵
    x 一维样本数组
    n 一维样本个数/阶数
    u 一维样本均值
    return 协方差矩阵
    """
    X = np.mat(np.zeros(n)).T
    for i in range(n):
        X[i, 0] = x[i]
    return (X-u)*(X-u).T
```

```python
C_x = get_C(x, n, u)
print("协方差矩阵为：\n", C_x)
```

## （2）计算特征值和特征向量

```python
eva, fec = np.linalg.eig(C_x)
print("特征值为：\n", eva, "\n特征向量为：\n", fec)
```

```python
print("上述特征值排序后的数组下标为：", np.argsort(eva))
```
