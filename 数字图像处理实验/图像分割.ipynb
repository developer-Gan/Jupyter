## 读入图片并计算灰度直方图

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_origin = cv2.imread("OpenCV.png", 0)
plt.imshow(img_origin, cmap = "gray")
print(img_origin.shape)
plt.show()
```

计算灰度值

```python
def get_hist(img):
    """
    获取直方图数组
    :param img: 输入图像
    :return: 直方图数组
    """
    b = 0
    hist = np.zeros(256, dtype=np.uint8)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            hist[img.item(i, j)] += 1
            print(i, j, img.item(i, j), hist[img.item(i, j)])
            b += 1
            print(b)
    return hist
```

```python
hist = get_hist(img_origin)
```
