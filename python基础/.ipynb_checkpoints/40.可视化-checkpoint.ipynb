## 改变图的属性

设置“点”的类型 -- 在 plt.plot() 中增加第三个实参的取值，如“o”

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.plot(np.arange(10), np.arange(10), "o")
plt.show()
```

设置线的颜色与形状 -- 修改 plt.plot() 的第三个实参

```python
plt.plot(np.arange(10), np.arange(10), "g--")
plt.show()
```

```python
plt.plot(np.arange(10), np.arange(10), "rD") # "rD" 的含义为 红色 + 钻石
plt.show()
```

显示汉字的方法：用 plt.rcParams() 设置汉字字体 应在 plt.plot() 调用之前使用

图名，X/Y 轴的函数必须在 plot() 和 show() 之间

```python
plt.rcParams["font.family"] = "SimHei" # 常用汉字字体有 SimHei、Kaiti、Lisu、Fangsong、YouYuan等
plt.plot(np.arange(10), np.arange(10), "r--")
plt.title("此处为图名") # 设置图名的函数
plt.xlabel("x轴的名称") # 设置修改X轴名称
plt.ylabel("y轴的名称") # 设置修改Y轴名称
plt.show()
```

图例的位置： plt.legend(loc = "位置")

```python
women = pd.read_csv("women.csv")
plt.rcParams["font.family"] = "SimHei"
plt.plot(women["height"],women["weight"], "g--")
plt.title("此处为图名")
plt.xlabel("x轴的名称") 
plt.ylabel("y轴的名称") 

plt.legend(loc = "upper left")
plt.show()
```

## 改变图的坐标轴的取值范围

横坐标和纵坐标的定义：分别采用 plt.xlim() 和 plt.ylim()

横纵坐标同时定义：plt.axis()

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(0, 10, 100) # 返回一个含有100个元素且每个元素的取值范围为 [0,10] 的等距离数列
x
```

```python
plt.plot(x, np.sin(x))
plt.xlim(11, -2) # x 取值范围[11, -2]
plt.ylim(2.2, -1.3) # y 取值范围[2.2, -1.3]
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 21, -1.6, 1.6])
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 21, -1.6, 1.6])
plt.axis("equal") # 此语句含义为X轴Y轴的刻度单位一样
plt.show()
```

## 去掉边界的空白

去掉边界空白的方法为： plt.axis("tight")

```python
import numpy as np
import matplotlib.pyplot as plt
plt.plot(x, np.sin(x))
plt.axis([-1, 21, -1.6, 1.6])
plt.axis("tight")
plt.show()
```

## 在同一个坐标上画两个图

分别定义 label ， 最后一个 plt.legend() 显示多个labels

```python
plt.plot(x, np.sin(x), label = "sin(x)")
plt.plot(x, np.cos(x), label = "cos(x)")
plt.axis("equal")
plt.legend() # 如果没有此行， label名称将无法显示
```

```python
plt.plot(x, np.sin(x), label = "sin(x)")
plt.plot(x, np.cos(x), label = "cos(x)")
```

## 多图显示

每个 plot 之前加一个 subplot(x, y, z)

x:行数
    
y:列数

z：接下来的 plot() 的画图位置，即放在第几个子窗口中

```python
plt.subplot(2, 3, 5), plt.scatter(women["height"], women["weight"]) 
plt.subplot(2, 3, 1), plt.scatter(women["height"], women["weight"])
plt.show()
```

## 散点图的画法

先生成即将用于可视化的试验数据集 X 和 Y

生成方法： python 的 sklearn.datasets.samples_generator 模块提供的函数 make_blobs()

生成符合正态分布的随机数据集

```python
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt
X, y = make_blobs(n_samples = 300, centers = 4, random_state = 0, cluster_std = 1.0) 
# n_samples: 样本数量
# n_features: 每个样本的特征数量，即列数
# centers: 类别数
# random_state: 随机数的生成方式
# cluster_std: 每个类别的方差
# 返回值有两个：
# X: 测试集，类型为数组，形状为[n_samples, n_features]
# y: 每个成员的标签（label）， 也是个数组，形状为[n_samples]的数组

plt.scatter(X[:, 0], X[:, 1], c = y, s = 50, cmap = "rainbow")
# X[:, 0] 和 X[:, 1] 分别为x坐标和y坐标
# c 为颜色
# s 为点的大小
# cmap 为色带，是c的补充
```

## pandas 可视化

在数据分析与数据科学项目中，还可以直接使用 pandas 的画图函数，它集成和优化了 Matplotlib，使 DateFrame 类数据的可视化处理更容易

```python
import pandas as pd
women = pd.read_csv("women.csv")
women.plot(kind = "bar") # king 决定图的类型
plt.show() # 未输入不会显示图标，只给出提示
```

```python
women.plot(kind = "barh") # 代表横向柱状图
plt.show()
```

```python
women.plot(kind = "bar" , x = "height", y = "weight", color = "g")
plt.show()
```

```python
women.plot(kind = "kde") # 代表的使核密度估计曲线
plt.show()
```

```python
women.plot(kind = "bar", x = "height", y = "weight", color = "g")
plt.legend(loc = "best") # 图里位置为“最优”
plt.show()
```

## seaborn 可视化

seaborn 可以实现比 matplotlib 更美观的可视化效果，需要与 matplotlib 结合使用

```python
import matplotlib.pyplot as plt
plt.style.use("classic")
%matplotlib inline
import numpy as np
import pandas as pd
```

数据准备

```python
Rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500) # 生成500个0-10的数
y = np.cumsum(Rng.randn(500, 6), 0)
# .cumsum 是 matlab 中的一个函数，通常用于计算一个数组各行的累加值
# 语法为： B = cumsum(A, dim), 或 B = cumsum(A)
```

matplotlib 的可视化效果往往并不美观

```python
plt.plot(x, y)
plt.legend("abcdef", ncol = 2, loc = "upper left")
```

用 seaborn 更美观

```python
import seaborn as sns
sns.set()
plt.plot(x, y)
plt.legend("abcdef", ncol = 2, loc = "upper left")
```

```python
sns.kdeplot(women.height, shade = True) # 核密度估计图
```

```python
sns.distplot(women.height) # 绘制 displot 图。功能为“直方图 + kdeplot”
```

```python
sns.pairplot(women) # 绘制散点图矩阵
```

函数 sns.jointplot() 的功能为绘制“联合分布图”

```python
sns.jointplot(women.height, women.weight, kind = "reg")
```

可以加一个 with 语句，设置美观，后面有一个冒号：

```python
with sns.axes_style("white"):
    sns.jointplot(women.height, women.weight, kind = "reg")
```

seaborn 还可以放在 for 语句中， 将多个变量画在一起

```python
for x in ["height", "weight"]:
    plt.hist(women[x], normed = True, alpha = 0.5) # 绘制直方图
```

（3）可视化绘图

```python
sns.set_style("darkgrid") # 设置 seanborn 的绘图样式或主题为 darkgrid （灰色 + 网格）
sns.stripplot(data = salaries, x = "rank", y = "salary", jitter = True, alpha = 0.5) 
# data 为数据来源；x 和 y 分别用于设置X轴和Y轴；jitter 的含义是散点是否有抖动； alpha 为透明度

sns.boxplot(data = salaries, x = "rank", y = "salary") # 继续绘制箱线图
```
